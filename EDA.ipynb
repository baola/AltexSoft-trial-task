```python
train_data.info()
```

```python
for i in range(6):
    for letter in ['u','g','i','r','z']:
        column_name='{}_{}'.format(letter,i)
        train_data[column_name] = train_data[column_name].apply(pd.to_numeric, downcast='float', errors='coerce')
    
for letter in ['u','g','i','r','z']:
        column_name='{}_6'.format(letter)
        train_data[column_name] = train_data[column_name].apply(pd.to_numeric, downcast='signed', errors='coerce')
for letter in ['rowv','colv']:
        train_data[letter] = train_data[column_name].apply(pd.to_numeric, downcast='float', errors='coerce')
```
